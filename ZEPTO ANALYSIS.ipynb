```python
## Loading the excel file 
```

```python
import pandas as pd

df=pd.read_excel("zepto orders Mycaptain.xlsx") 
df
```

```python
##Converting date columns
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"]) 
df["Delivery_Time"] = pd.to_datetime(df["Delivery_Time"])

df["Order_Date"]
df["Delivery_Time"]
```

```python
##Calculating delivery duration in minutes
```

```python
df["Delivery_Duration"] = (df["Delivery_Time"] - df["Order_Date"]).dt.total_seconds() / 60
df["Delivery_Duration"]
```

```python
##Average delivery time for each city
```

```python
avg_city = df.groupby("City")["Delivery_Duration"].mean().reset_index()
print(avg_city) 
```

```python
##Average customer rating for each category
```

```python
avg_rating = df.groupby("Category")["Customer_Rating"].mean().reset_index() 
print(avg_rating) 
```

```python
## Total revenue (total order value) for each category
```

```python
revenue = df.groupby("Category")["Order_Value"].mean().reset_index()
print(revenue)
```

```python
##Are Express deliveries faster than Normal deliveries
```

```python
mode_compare = df.groupby("Delivery_Mode")["Delivery_Duration"].mean().reset_index() 
print(mode_compare)
```

```python
##Average delivery fee for each category
```

```python
avg_fee = df.groupby("Category")["Delivery_Fee"].mean().reset_index()
print(avg_fee)
```

```python
###Total delivery fee collected by each city
```

```python
city_fees = df.groupby("City")["Delivery_Fee"].mean().reset_index()
print(city_fees)
```

```python
## Average customer rating by delivery mode
```

```python
rating_mode = df.groupby("Delivery_Mode")["Customer_Rating"].mean().reset_index()
print(rating_mode)
```

```python
 ##Top-selling category
```

```python
revenue_cat = df.groupby("Category")["Order_Value"].sum().reset_index() 
top_selling = revenue_cat.sort_values(by="Order_Value", ascending=False)

print(top_selling)
```

```python
##Fastest and slowest categories 
```

```python
speed_cat = df.groupby("Category")["Delivery_Duration"].mean().reset_index() 
sorted_speed = speed_cat.sort_values(by="Delivery_Duration")

print(sorted_speed)
```
